```python
import numpy as np
from sklearn import model_selection
```

```python
def step_gradient(points,learning_rate,m,c):
  m_slope = 0
  c_slope = 0
  N = len(points)
  for i in range(N):
    x = points[i,0]
    y = points[i,1]
    m_slope += (-2/N)*(y-m*x -c)*x
    c_slope += (-2/N)*(y-m*x-c)
  new_m = m-learning_rate*m_slope
  new_c = c-learning_rate*c_slope
  return new_m,new_c

```

```python
#gd(): Gradient Descent Function
def gd(points,learning_rate,num_iterations):
    m = 0
    c = 0
    for i in range(num_iterations):
      m,c = step_gradient(points,learning_rate,m,c) 
      print(i,"Cost: ",cost(points,m,c))
    return m,c
```

```python
def cost(points,m,c):
  total_cost = 0
  N = len(points)
  for i in range(N):
    x = points[i,0]
    y = points[i,1]
    total_cost += (1/N)*((y-m*x-c)**2)
  return total_cost
```

```python
def run():
  data = np.loadtxt("data.csv",delimiter=",")
  learning_rate = 0.0001                            #random value of alpha hit and trial method to find best alpha
  num_iterations = 12
  m,c = gd(data,learning_rate,num_iterations)
  print(m,c)
```

```python
run()
```
